```python

from astropy.io import ascii

IFU = "044"

fncatalog = "catalog_{}.txt".format(IFU)
fnout = "catalog_{}.reg".format(IFU)

h="""# Region file format: DS9 version 4.1
global color=green dashlist=8 3 width=1 font="helvetica 10 normal roman" select=1 highlite=1 dash=0 fixed=0 edit=1 move=1 delete=1 include=1 source=1
fk5
"""

t = ascii.read(fncatalog)
with open(fnout,'w') as f:
    f.write(h)
    for r in t:
        s = 'circle({},{},{}") # color=grey font="helvetica 6 normal roman" text={{{}:{}}}\n'.format(r["ra_com"],r["dec_com"],r["diam"], r["id"],r["wl_com"])
        f.write(s)
```

# convert to luminosity

```python
from astropy.cosmology import WMAP9 as cosmo
from astropy import units as u
import numpy as np

def toL(f,z):
    
    DL = cosmo.luminosity_distance(z).to(u.cm) # luniosity distance in cm
    print(DL.to(u.Mpc))
    return 4.*np.pi*DL**2. * f * u.erg / u.second / (u.cm)**2.

# HDLAB1
toL(f=1.e-16,z=2.16)

toL(f=6.1e-17,z=2.19)

#HTUFO1
toL(f=3.02e-17,z=2.51)
```

```python
def toKpc(size,z):
    DL = cosmo.angular_diameter_distance(z) # Angular diameter distance in kpc
    print(DL)
    print("z = {:.2f}, {:.1f} ".format(z, DL.to(u.kpc) * (4.85e-6  ) ))
    return DL.to(u.kpc) * np.deg2rad( size/3600. )

print("HDLAB1 Extent = ", toKpc(4.7956*2.,z=2.16))
print("HDLAB2 Extent = ", toKpc(8.6,z=2.19))

print("HDFIL1 Extent = ", toKpc(19.2,z=2.7))
print("HDFIL2 Extent = ", toKpc(40.,z=2.4))

print("HDFIL4 Extent = ", toKpc(20.,z=2.96))


```

# Dynamical Mass estimate

```python

s_e = 19.86/3842.52 * 3e5

scale = 8.441 #kpc/arcsec from ned wrights comso cal at z = 2.16

R_e = np.sqrt( 5.**2. + 3.5*2. ) * scale

G = 4.3e-9 # Mpc M_sun^-1 (km/s)^-2

M = s_e**2.*(R_e/1000.)/G

print("estimate M_dyn = {:.2e} M_sun".format(M))
```
